```python
custom_module = 'savasy/bert-base-turkish-ner-cased'
ner = pipeline('ner', model = custom_module, tokenizer = custom_module)
sequence = "Merhaba! Benim adin Sinan. San Francisco'dan geliyorum"
ner(sequence)
```

```python
model_name = "deepset/roberta-base-squad2"
qa = pipeline(model = model_name, tokenizer = model_name, revision = "v1.0", task = "question-answering")
```

```python
sequence = "Where is Saket living these days?", "Saket lives in Gurugram but Jasuja lives in Ganganagar"
sequence
```

```python
qa(*sequence)
```
